## ChatOpenAI

```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model='gpt-4o')
```

```python
llm
```

#### Input & Response from LLM

```python
response = llm.invoke('Hi, OpenAI!')
```

```python
response
```

### ChatPromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate([
    ('system','You are an export AI Engineer. Answer my questions.'),
    ('user','{input}')
])
```

```python
prompt
```

### Chaining Response

```python
chain = prompt | llm
```

```python
bot_response = chain.invoke({'input':'Tell me about Langsmith.'})
```

```python
bot_response
```

```python
type(bot_response)
```

### StrOutputParser

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
output_parser = StrOutputParser()
```

```python
chain = prompt|llm|output_parser
```

```python
bot_response = chain.invoke({'input':'Tell me about Langsmith.'})
```

```python
bot_response
```
